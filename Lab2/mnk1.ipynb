```python
# imoort cell
import numpy as np
```

```python
def MNK_first():
    alpha0 = 1. #mortality rate due to predators
    beta = 1.
    delta = 1.
    omega = 0.
    gamma = 1.
    x0 = 4
    y0 = 2
    X0 = (x0, y0)

    Nt = 1000
    tmax = 30
    t_array = np.linspace(0.,tmax, Nt)
    dt = t_array[1] - t_array[0]

    nt = len(t_array)
    X  = np.zeros([nt, len(X0)])
    X[0] = X0

    for i in range(0, Nt - 1):
        k1 = 1 - alpha0 * (1 + np.sin(omega * (i + 1) * dt)) * dt
        k2 = 1 + delta * dt

        c1 = k2 * dt
        c2 = k2 * k1 - beta * dt * X[i][0] - X[i][1] * dt
        c3 = -1. * X[i][1] * k1

        discriminant = c2 ** 2 - 4 * c1 * c3
        X[i + 1][1] = (-1. * c2 + np.sqrt((discriminant))) / (2. * c1)
        X[i + 1][0] = X[i][0] / (k1 + X[i + 1][1] * dt)

    return t_array, X
```

```python
t, answer = MNK_first()
```

```python
import matplotlib.pyplot as plt
plt.figure()
plt.grid()
plt.title("MNK")
plt.plot(t, answer[:, 0], 'xb', label = 'Victims')
plt.plot(t, answer[:, 1], '+r', label = "Predators")
plt.xlabel('Time t, [days]')
plt.ylabel('Population')
plt.legend()

plt.show()
```

```python
plt.figure()
plt.plot(answer[:, 0], answer[:, 1], "-")
plt.xlabel("Victims")
plt.ylabel("Predators")
plt.grid()
plt.title("Phase plane : Victims vs Predators (MNK3)")
```
